```python
# Let's gather our essential tools for this quest!
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

**Chapter 1: Whispers in the Meadow â€“ Gathering Data**

To begin her quest, BiddyBot listened to the worldâ€™s whispers.
She carefully observed the tiny data fireflies fluttering through the meadow classroom.
Each firefly carried information:  
âœ¨ one spark for **study hours (x)**  
âœ¨ one spark for **exam score (y)**.



```python
# BiddyBot's Code for Chapter 1
np.random.seed(42)
study_hours = np.random.rand(50, 1) * 10
exam_scores = 2 * study_hours + np.random.randn(50, 1) * 2 + 50

plt.figure(figsize=(9, 6))
plt.scatter(study_hours, exam_scores, color='blue', alpha=0.7, label='Observed Data')
plt.title("Chapter 1: BiddyBot's Collected Data Points")
plt.xlabel('Study Hours (x)')
plt.ylabel('Exam Scores (y)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

**Chapter 2: The Magical Thread â€“ Finding the Best Fit Line**

With her data collected, BiddyBot felt a golden thread appear â€”
the *Line of Best Fit* weaving through her data.
Her goal: position it so it was as close as possible to every data firefly,
minimizing the tiny residuals between each point and the path.

![](https://raw.githubusercontent.com/deeplearningangel/BiddyBot-Linear-Regression-Story/refs/heads/main/Original/P1.png)

```python
# Train Linear Regression
model = LinearRegression()
model.fit(study_hours, exam_scores)

m = model.coef_[0][0]
b = model.intercept_[0]

predicted_scores = model.predict(study_hours)

plt.figure(figsize=(9, 6))
plt.scatter(study_hours, exam_scores, color='blue', alpha=0.7, label='Observed Data')
plt.plot(study_hours, predicted_scores, color='red', linewidth=3, label="BiddyBot's Best-Fit Line")
for i in range(len(study_hours)):
    plt.plot([study_hours[i], study_hours[i]],
             [exam_scores[i], predicted_scores[i]], 'k--', lw=0.6, alpha=0.4)
plt.title("Chapter 2: BiddyBot's Line of Best Fit with Errors")
plt.xlabel('Study Hours (x)')
plt.ylabel('Exam Scores (y)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

![](https://raw.githubusercontent.com/deeplearningangel/BiddyBot-Linear-Regression-Story/refs/heads/main/extra/P3_1.png)

```python
new_study_hours = np.array([[7.5]])
predicted_new_score = model.predict(new_study_hours)

plt.figure(figsize=(9, 6))
plt.scatter(study_hours, exam_scores, color='blue', alpha=0.7, label='Observed Data')
plt.plot(study_hours, predicted_scores, color='red', linewidth=3, label='Regression Line')
plt.scatter(new_study_hours, predicted_new_score,
            color='green', s=200, marker='X', edgecolors='white', linewidth=2,
            label=f'Prediction for {new_study_hours[0,0]:.1f} hrs')

x_pred, y_pred = new_study_hours[0,0], predicted_new_score[0,0]
plt.plot([x_pred, x_pred], [plt.gca().get_ylim()[0], y_pred], 'g--', lw=1.8)
plt.plot([plt.gca().get_xlim()[0], x_pred], [y_pred, y_pred], 'g--', lw=1.8)

plt.title("Chapter 3: BiddyBot's Prediction in Action")
plt.xlabel('Study Hours (x)')
plt.ylabel('Exam Scores (y)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()

print(f"If a student studies for {x_pred:.1f} hours, BiddyBot predicts a score of {y_pred:.1f}.")
```

```python
print(f"Slope (m): {m:.2f}")
print(f"Intercept (b): {b:.2f}")
print(f"Equation: y = {m:.2f}x + {b:.2f}")
```

```python
# Quick evaluation example
X_train, X_test, y_train, y_test = train_test_split(study_hours, exam_scores,
                                                    test_size=0.2, random_state=42)
model_eval = LinearRegression()
model_eval.fit(X_train, y_train)
y_pred = model_eval.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("ðŸ“˜ BiddyBot's Evaluation Report")
print(f"Mean Squared Error: {mse:.2f}")
print(f"R-squared Score: {r2:.2f}")
print("\nBiddyBot smiles â€” her model performs well, but the quest for knowledge never ends!")
```
